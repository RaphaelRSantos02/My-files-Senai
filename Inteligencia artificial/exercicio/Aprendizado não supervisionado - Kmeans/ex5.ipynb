```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dados = np.array([
    [70.2, 12.5, 4.7],
    [65.1, 8.2, 3.9],
    [75.5, 15.6, 5.1],
    [80.3, 10.2, 4.5],
    [68.7, 11.8, 4.2],
    [72.9, 14.3, 5.3],
    [78.6, 9.8, 4.8],
    [66.4, 8.9, 4.0],
    [73.1, 13.7, 5.0],
    [69.5, 12.1, 4.3]
])
```

```python
num_clusters = 3  
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(dados)
```

```python
df = pd.DataFrame(dados, columns=['Temperatura (°C)', 'Vibração', 'Corrente'])
df['Cluster'] = clusters
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Temperatura (°C)', data=df, palette='Set1')
plt.title('Boxplot da Temperatura')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Vibração', data=df, palette='Set2')
plt.title('Boxplot da Vibração')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Corrente', data=df, palette='Set3')
plt.title('Boxplot da Corrente')
plt.show()
```
