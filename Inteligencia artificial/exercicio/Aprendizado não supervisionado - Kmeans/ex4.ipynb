```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

```

```python
dados = {
    'Substância': ['Álcool', 'Gasolina', 'Leite', 'Querosene', 'Óleo', 'Vinho'],
    'Concentração (%)': [12.5, 0.1, 4.0, 1.2, 0.5, 15.0],
    'Teor Alcoólico (%)': [50, 0.05, 0.01, 0.02, 0.01, 12.5]
}

df = pd.DataFrame(dados)
```

```python
X = df[['Concentração (%)', 'Teor Alcoólico (%)']]

num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

df['Cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(8, 6))


for cluster in range(num_clusters):
    cluster_data = df[df['Cluster'] == cluster]
    plt.scatter(cluster_data['Concentração (%)'], cluster_data['Teor Alcoólico (%)'], label=f'Cluster {cluster}')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', label='Centroids')
plt.xlabel('Concentração (%)')
plt.ylabel('Teor Alcoólico (%)')
plt.title('Agrupamento das Substâncias')
plt.legend()
plt.grid(True)
plt.show()
```
