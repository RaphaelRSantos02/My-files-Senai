```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import tree
```

```python
iris = datasets.load_iris()
```

```python
data = iris.data
target = iris.target
```

```python
X,y = iris.data, iris.target
```

```python
classificador = tree.DecisionTreeClassifier() #instacio o modelo da arvore de deciso
fig = plt.figure(figsize=(30,25))
classificador.fit(X,y) #treino o modelo
tree.plot_tree(classificador,
               feature_names= iris.feature_names,class_names= iris.target_names.tolist()
               ,filled=True)
plt.savefig('arvore1.jpg')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(iris.data[:,0],iris.data[:,2],
            c=iris.target)
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[2])
plt.tight_layout()
plt.show()
```
